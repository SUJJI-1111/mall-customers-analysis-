```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from google.colab import files
import pandas as pd

# 1. Upload the file (correct filename)
uploaded = files.upload()  # Make sure to select "Mall_Customers.csv"

# 2. Load the data (verify exact filename after upload)
df = pd.read_csv("Mall_Customers.csv")  # Case-sensitive!

# --- Data Cleaning Steps ---
# 1. Standardize gender (Male/Female) - FIXED SYNTAX
df['Gender'] = df['Gender'].str.lower().replace({'male': 'male', 'female': 'female'})

# 2. Remove duplicates - CORRECTED
df.drop_duplicates(subset=['CustomerID'], keep='first', inplace=True)

# 3.Validation check:
assert df.isnull().sum().sum() == 0, "Dataset contains missing values!"

# 4. Save cleaned data
df.to_csv("cleaned_mall_customers.csv", index=False)

# 5. Display results - CLEAN OUTPUT
print("\n=== Cleaned Data ===")
print(f"Total rows: {len(df)}")  # Should be 200
pd.set_option('display.max_rows', 10)  # Shows 10 rows (5 head + 5 tail)
display(df)  # Clean notebook  display

plt.scatter(df['Age'], df['Spending Score (1-100)'], alpha=0.5, color='blue')
plt.show()

fig = px.scatter(df, x='Age', y='Spending Score (1-100)')
fig.show()
```
